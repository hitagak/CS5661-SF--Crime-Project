```python
import numpy as np
import pandas as pd

import time
import datetime

from collections import Counter

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# read the dataset and generate dataframe
crime_df = pd.read_csv('data_20000.csv')
crime_df.head()
```

#### Convert Date to milliseconds

```python
# convert date to timestamp (milliseconds, divide by a million to get a more readable decimal)
crime_df['Dates'] = crime_df['Dates'].apply(lambda x: time.mktime(time.strptime(x, '%m/%d/%Y %H:%M'))/1000000)

crime_df.head()
```

#### get Label Column

```python
new_crime_df = crime_df.loc[:,['Category'] ]

new_crime_df.head()

```

#### Label Hot Encoding

```python
my_label= ['Category']

new_crime_df=pd.get_dummies(new_crime_df,columns=['Category'],
                  drop_first=False)

new_label = list(new_crime_df.dtypes.index)

y = new_crime_df[new_label]

print(y)

print(new_label)

print("\nTotal category: ", len(new_label))
```

#### Features Hot Encoding

```python
my_list = ['Dates','Descript','DayOfWeek','PdDistrict','Resolution','Address','X','Y','Category']



# split sentences into series of words so each word can be a column
def string_to_series_of_words(x):    
    return pd.Series(x.split(" "))


for f in my_list[1:6]:
    crime_df[f] = crime_df[f].apply(string_to_series_of_words)

    
crime_df=pd.get_dummies(crime_df,columns=['Descript','DayOfWeek','PdDistrict','Resolution','Address'],
                  drop_first=False)

crime_df.head()


new_features = list(crime_df.dtypes.index)
new_features.remove('Category')

X = crime_df[new_features]

print(X)

print(new_features)

print("\nTotal features: ", len(new_features))
```

#### split the data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=1)
```

#### DecisionTree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

decTree =  DecisionTreeClassifier(criterion='entropy', min_samples_split=4, random_state=0)
decTree.fit(X_train, y_train)
y_predict_decTree = decTree.predict(X_test)

accuracy_decTree = accuracy_score(y_test, y_predict_decTree)

print("\nAccuracy Decision Tree: ", accuracy_decTree) 
```

#### Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

rf =  RandomForestClassifier(n_estimators = 50, bootstrap = True, random_state=0, criterion='entropy')
rf.fit(X_train, y_train)

y_predict_randF = rf.predict(X_test)

accuracy = accuracy_score(y_test, y_predict_randF)

print("\nAccuracy: ", accuracy) 
```

#### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression


logReg = LogisticRegression()
logReg.fit(X_train, y_train)

y_predict = logReg.predict(X_test)

#print(y_predict)
#print(y_test) 

accuracy = accuracy_score(y_test, y_predict)

#print("\nAccuracy: ", accuracy)
```

#### ANN

```python
from sklearn.neural_network import MLPClassifier

# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=2)


# my_ANN is an instance of MLPClassifier class
my_ANN = MLPClassifier(hidden_layer_sizes=(80,), activation= 'logistic', solver='adam', alpha=1e-5, 
                       random_state=1, learning_rate_init = 0.002)

# training
my_ANN.fit(X_train, y_train)

# predict label
ann_y_predict = my_ANN.predict(X_test)

# compare the result between actual label and predicted label
accuracy = accuracy_score(y_test, ann_y_predict)

print("Accuracy of MLPclassifier :", accuracy)
```

#### Ada boost

```python
#Ada boost
from sklearn.ensemble import AdaBoostClassifier
dt = DecisionTreeClassifier()
clf = AdaBoostClassifier(n_estimators=100, base_estimator=dt,learning_rate=1)
clf.fit(X_train, y_train)
prediction = clf.predict(X_test)
acc_score = accuracy_score(y_test, prediction)
print("Accuracy by Ada Boost Classifier: ", acc_score)
```

#### KNN

```python
from sklearn.neighbors import KNeighborsClassifier

k=500

my_knn = KNeighborsClassifier(n_neighbors=k)
my_knn.fit(X_train, y_train)
y_predict = my_knn.predict(X_test)
accuracyKnn = accuracy_score(y_test, y_predict)
print("Accuracy using KNN", accuracyKnn)
```
